```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset=dataset.drop("User ID",axis=1)

dataset
```

```python
dataset=dataset.drop("Age",axis=1)

dataset
```

```python
dataset=dataset.drop("Gender_Male",axis=1)

dataset
```

```python
indep=dataset[['EstimatedSalary']]
indep
```

```python
dep=dataset[['Purchased']]
dep
```

```python
dataset["Purchased"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(indep,dep,test_size=1/3,random_state=0)
x_train,x_test,y_train,y_test
```

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LinearRegression 
from sklearn.isotonic import IsotonicRegression
from sklearn.model_selection import GridSearchCV
param_grid={'increasing':['bool','auto'],'out_of_bounds':['nan','clip','raise']}      
grid=GridSearchCV(IsotonicRegression(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_
grid_predictions=grid.predict(x_test)
grid_predictions
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid_predictions)
print("Confusion Matrix:\n",cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,grid_predictions)
print("classification_report:\n",clf_report)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_predictions,average='weighted')
print("The f1_macro value for best parameter{}:\n".format(grid.best_params_),f1_macro)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid_predictions)
```
